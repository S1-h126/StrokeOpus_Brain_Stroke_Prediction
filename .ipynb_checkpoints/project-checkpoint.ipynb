```python
target = 'stroke'
X = all_data.loc[:,all_data.columns!=target]
y = all_data.loc[:,all_data.columns==target]
```

```python
cols = all_data[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']]
import seaborn as sns
plt.figure(figsize=(20,30), facecolor='grey')
plotnumber=1

for column in cols:
    if plotnumber<=15:
        ax = plt.subplot(5,3,plotnumber)
        sns.kdeplot(cols[column])
        plt.xlabel(column, fontsize=20)
        
    plotnumber+=1
plt.tight_layout()
```

```python
X = pd.get_dummies(X)
```

```python
ax = sns.countplot(x=target,data=all_data)
print(all_data[target].value_counts())
```

```python
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
ay = sns.countplot(x=target,data=y)
print(y[target].value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train,Y_test = train_test_split(X,y, test_size=0.20, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

def rf(X_train, Y_train, n_estimators=300):
 rndforest = RandomForestClassifier(n_estimators=n_estimators, n_jobs=-1)
 rndforest.fit(X_train, Y_train)
 return rndforest
```

```python
random_forest_en = rf(X_train,Y_train.values.ravel(), n_estimators=300)
```

```python
#printing the accuracy
print("TRAINING ACCURACY:", random_forest_en.score(X_train, Y_train))
print("TESTING ACCURACY", random_forest_en.score(X_test, Y_test))
```

```python
#saving the model
import pickle

# Save the trained model
with open('deploy.pkl', 'wb') as model_file:
    pickle.dump(random_forest_en, model_file)
```

```python
with open('deploy.pkl', 'rb') as model_file:
    model = pickle.load(model_file)
```
